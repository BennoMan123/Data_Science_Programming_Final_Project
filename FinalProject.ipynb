## How many and what percentage of the goals scored were from special teams? (from short handed and power play goals)

```python
totalGoals = int(stats[('Shots', 'G')].sum())
print(f"There were {totalGoals} total goals scored so far this season.")
```

```python
totalSHG = int(stats[('Goals', 'SHG')].sum())
print(f"There was {totalSHG} short handed goal scored so far this season.")
```

```python
totalPPG = int(stats[('Goals', 'PPG')].sum())
print(f"There were {totalPPG} power play goals scored so far this season.")
```

```python
totalEVG = totalGoals - totalPPG - totalSHG
print(f"There were {totalEVG} even strength goals scored so far this season.")
```

```python
goalDict = {
    f"Short Handed Goals ({totalSHG}/{totalGoals})" : totalSHG/totalGoals,
    f"Power Play Goals ({totalPPG}/{totalGoals})" : totalPPG/totalGoals,
    f"Even Strength Goals ({totalEVG}/{totalGoals})" : totalEVG/totalGoals
}
```

```python
plt.pie(
    goalDict.values(),
    labels= goalDict.keys(),
    autopct='%.2f%%',
    colors = ["slateblue", "olivedrab", "orange"]
)
plt.title("Percentage of Each Type of Goal")
plt.show()
```

### About 13.04% of goals (6 out of 46 goals) were power play goals and 1 goal was a short handed goal making up about 2.17% of goals.
